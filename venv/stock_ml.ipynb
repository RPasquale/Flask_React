```python
import yfinance as yf
import pandas as pd

symbol = 'AAPL'  # Replace with the stock symbol you're interested in

# Get stock information
stock = yf.Ticker(symbol)

# Get company overview
overview_data = stock.info
overview_df = pd.DataFrame.from_dict(overview_data, orient='index').T

# Get historical 
historical_data = stock.history(period='max')
historical_df = pd.DataFrame(historical_data)

# Get historical Income Statement
income_data = stock.quarterly_income_stmt.T
income_df = pd.DataFrame(income_data)

# Get historical balance sheet
balance_data = stock.quarterly_balance_sheet.T
balance_df = pd.DataFrame(balance_data)

# Get historical cash flow
cashflow_data = stock.quarterly_cashflow.T
cashflow_df = pd.DataFrame(cashflow_data)

# Print the dataframes
print("Company Overview:")
print(overview_df)

print("\nHistorical Income Statement:")
print(income_df)

print("\nHistorical Balance Sheet:")
print(balance_df)

print("\nHistorical Cash Flow:")
print(cashflow_df)
```

```python
#print(cashflow_df)
print(income_df)
print(cashflow_df)
print(balance_df)
```

```python
# Combine the DataFrames into a main DataFrame
main_df = pd.concat([income_df, balance_df, cashflow_df], axis=1)

# Print the main DataFrame
print(main_df)
```
