```python
import yfinance as yf
import pandas as pd

symbol = 'AAPL'  # Replace with the stock symbol you're interested in

# Get stock information
stock = yf.Ticker(symbol)

# Get company overview
overview_data = stock.info
overview_df = pd.DataFrame.from_dict(overview_data, orient='index').T

# Get historical 
historical_data = stock.history(period='max')
historical_df = pd.DataFrame(historical_data)

# Get historical Income Statement
income_data = stock.quarterly_income_stmt.T
income_df = pd.DataFrame(income_data)

# Get historical balance sheet
balance_data = stock.quarterly_balance_sheet.T
balance_df = pd.DataFrame(balance_data)

# Get historical cash flow
cashflow_data = stock.quarterly_cashflow.T
cashflow_df = pd.DataFrame(cashflow_data)

# Print the dataframes
print("Company Overview:")
print(overview_df)

print("\nHistorical Income Statement:")
print(income_df)

print("\nHistorical Balance Sheet:")
print(balance_df)

print("\nHistorical Cash Flow:")
print(cashflow_df)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.model_selection import train_test_split


# Shift the Close price column
aapl_historical['Close(t-1)'] = aapl_historical['Close'].shift(1)

# Remove the first row
aapl_historical = aapl_historical.dropna()

# Split the adjusted data into features (X) and target variable (y)
X = aapl_historical.drop('Close', axis=1)  # Features
y = aapl_historical['Close']  # Target variable

# Split the adjusted data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1)

# Create an instance of the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(X_train)

# Transform the training and test data using the scaler
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt

neurons_range = range(1, 41, 4)
test_errors = []

for n in neurons_range:
    model = MLPRegressor(hidden_layer_sizes=(n),
                         activation='relu',
                         solver='adam',
                         alpha=0.001)
    model.fit(X_train, y_train)
    error = 1 - model.score(X_test, y_test)
    test_errors.append(error)
    print(f'Number of neurons: {n}, Test error: {error:.6f}')

min_error = min(test_errors)
min_index = test_errors.index(min_error)
best_neuron_count = neurons_range[min_index]

plt.plot(neurons_range, test_errors, marker='o')
plt.xlabel('Number of Neurons')
plt.ylabel('Test Error')
plt.title('Test Error vs. Number of Neurons')
plt.xticks(neurons_range)
plt.grid(True)
plt.scatter(best_neuron_count, min_error, color='red', label='Best Model')
plt.legend()
plt.show()

print(f"Smallest Test Error: {min_error:.6f}")
print(f"Number of Neurons: {best_neuron_count}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

neurons_range = range(1, 21, 2)
test_errors = []

for n in neurons_range:
    model = MLPRegressor(hidden_layer_sizes=(n),
                         max_iter = 1000,
                         tol=1e-4,
                         activation='relu',
                         solver='adam',
                         alpha=0.001)
    model.fit(X_train, y_train)
    error = 1 - model.score(X_test, y_test)
    test_errors.append(error)
    print(f'Number of neurons: {n}, Test error: {error:.6f}')

test_errors = np.array(test_errors)
mean_error = np.mean(test_errors)
std_error = np.std(test_errors)

# Set the threshold as a multiple of the standard deviation
threshold = 1  # Adjust this value as needed

# Filter out the outlier values
filtered_indices = np.where(np.abs(
    test_errors - mean_error) <= threshold * std_error)[0]  # Extract the array
filtered_neurons = [neurons_range[i] for i in filtered_indices]
filtered_errors = [test_errors[i] for i in filtered_indices]

min_error = min(filtered_errors)
min_index = filtered_errors.index(min_error)
best_neuron_count = filtered_neurons[min_index]

plt.plot(filtered_neurons, filtered_errors, marker='o')
plt.xlabel('Number of Neurons')
plt.ylabel('Test Error')
plt.title('Test Error vs. Number of Neurons (Filtered)')
plt.xticks(filtered_neurons)
plt.grid(True)
plt.scatter(best_neuron_count, min_error, color='red', label='Best Model')
plt.legend()
plt.show()

print(f"Smallest Test Error: {min_error:.6f}")
print(f"Number of Neurons: {best_neuron_count}")
```
